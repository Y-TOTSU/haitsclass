```python
import pandas as pd
df= pd.read_csv("train.csv")
```

```python
# 欠損データの確認
df.isnull().sum()
```

```python
# Embarked, Ageに欠損のあるサンプルを削除
df_1 = df.dropna(subset=["Embarked","Age"])
```

```python
# pandasでone-hotエンコーディングを使い、ダミー変数を作成
sex_dum = pd.get_dummies(df_1["Sex"])
emb_dum = pd.get_dummies(df_1["Embarked"])
```

```python
X = pd.concat([df_1.iloc[:,[2,5,6,7,9]], sex_dum, emb_dum], axis=1)
```

```python
y = df_1.iloc[:,1]
```

```python
X
```

```python
# ホールドアウト法による分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# ランダムフォレストによる学習
from sklearn.ensemble import RandomForestClassifier
rfc_1 = RandomForestClassifier(random_state=0, n_estimators=10)
rfc_1.fit(X_train, y_train)
```

```python
# 正答率を出力
print('train accuracy: %.3f' % rfc_1.score(X_train, y_train))
print('test  accuracy: %.3f' % rfc_1.score(X_test, y_test))
```
